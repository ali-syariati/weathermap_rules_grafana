## generate panel

```python
def panel_str(
        grid_h: int,
        grid_w: int,
        grid_x: int,
        grid_y: int,
        time_from: str,
        interval: str,
        title_opening: str,
        annotation: dict,
        template_dashboard: dict):
    template_dashboard["datasource"]["type"]=annotation['datasource_type']
    template_dashboard["datasource"]["uid"]=annotation['datasource_uid_1']
    template_dashboard["gridPos"]["h"] = grid_h
    template_dashboard["gridPos"]["w"] = grid_w
    template_dashboard["gridPos"]["x"] = grid_x
    template_dashboard["gridPos"]["y"] = grid_y
    template_dashboard["interval"] =  interval
    template_dashboard["targets"][0]["datasource"]["type"] = annotation['datasource_type']
    template_dashboard["targets"][0]["datasource"]["uid"]=annotation['datasource_uid_2']
    template_dashboard["targets"][0]["group"]["filter"] = annotation["group_filter"]
    template_dashboard["targets"][0]["host"]["filter"] = annotation["host_filter"]
    template_dashboard["targets"][0]["item"]["filter"] = annotation["item_filter_outbound"]
    template_dashboard["targets"][0]["itemTag"]["filter"] = annotation["item_tag_filter"]
    template_dashboard["targets"][1]["datasource"]["type"] = annotation['datasource_type']
    template_dashboard["targets"][1]["datasource"]["uid"]=annotation['datasource_uid_2']
    template_dashboard["targets"][1]["group"]["filter"] = annotation["group_filter"]
    template_dashboard["targets"][1]["host"]["filter"] = annotation["host_filter"]
    template_dashboard["targets"][1]["item"]["filter"] = annotation["item_filter_inbound"]
    template_dashboard["targets"][1]["itemTag"]["filter"] = annotation["item_tag_filter"]
    template_dashboard["timeFrom"] = time_from
    template_dashboard["title"] = f"{title_opening} {annotation['title']}"
    
    return template_dashboard
    
template_panel = {
    "refresh": "",
    "schemaVersion": 36,
    "style": "dark",
    "tags": [],
    "templating": {
        "list": []
    },
    "time": {
        "from": "now-5m",
        "to": "now"
    },
    "timepicker": {
        "hidden": True
    },
    "timezone": "",
} 

list_annotation_data = json.load(open("annotation/dashboard_query.json"))
for idx, annotation in enumerate(list_annotation_data):
    if idx == 0: continue
    
    file_name = re.sub(r'\W+', '_', annotation["title"])
    json_panel = json.load(open(f"dashboard_json/json_model/{file_name}.json"))
    
    json_panel["panels"].append(deepcopy(panel_str(9, 24, 0, 0, '24h', '5m', 'Daily (5 Minute Average)', annotation, template_dashboard_panel)))
    json_panel["panels"].append(deepcopy(panel_str(9, 24, 0, 9, '7d', '30m', 'Weekly (30 Minute Average)', annotation, template_dashboard_panel)))
    json_panel["panels"].append(deepcopy(panel_str(9, 24, 0, 18, '120d', '2h', 'Monthly (2 Hours Average)', annotation, template_dashboard_panel)))
    json_panel["panels"].append(deepcopy(panel_str(9, 27, 0, 27, '360d', '1d', 'Yearly (1 Day Average)', annotation, template_dashboard_panel)))
    
    for key, val in template_panel.items():
        json_panel[key] = val
    
    write_panel = open(f"dashboard_json/result_json/{file_name}.json", "w")
    write_panel.write(json.dumps(json_panel, indent=2))
    write_panel.close()
```

## generate annotation rules

```python
# generate annotation rules
result = []
for idx, annotation in enumerate(list_annotation_data):
    if idx == 0: continue
    
    host = annotation["host_filter"]
    item = " ".join(annotation["item_filter_outbound"].split(" ")[0:-2])
    iframe_link = annotation["link"].replace("10.54.68.73/d", "10.54.68.73/d-solo")
    rules = {
            "arrow_id_out": "",
            "label_id_out": "",
            "arrow_id_in": "",
            "label_id_in": "",
            "alias": host,
            "pattern_percentage_out": f"/{host}: {item} (.*speed|.*Percentage out)/",
            "pattern_percentage_in": f"/{host}: {item} (.*speed|.*Percentage in)/",
            "pattern_speed": f"{host}: {item} Link speed",
            "iframe": f"<iframe src=\"{iframe_link}&panelId=1\" width=\"450\" height=\"200\" frameborder=\"0\"></iframe>",
            "link_url": annotation["link"],

            "group_filter": annotation["group_filter"],
            "item_tag_filter": annotation["item_tag_filter"],
            "id_tag_filter": "",
            "item_filter": "/.*speed|.*Percentage.*/"
    }
    result.append(deepcopy(rules))
write_panel = open(f"annotation/generate_annotation_rules.json", "w")
write_panel.write(json.dumps(result, indent=2))
write_panel.close()
```
